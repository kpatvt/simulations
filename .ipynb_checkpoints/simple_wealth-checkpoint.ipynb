```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
# Size of the population
POPULATION = 100

# Let's see what happens
# Create the agents
agents = np.zeros(POPULATION)
agents_order = np.arange(POPULATION)

# Everyone has the same amount of wealth
agents[:] = 1.0

def step(agents):
    # Agents is modified in place
    # The population is the length of the agents
    pop = len(agents)
    
    # Get a random order agents
    random_agents_order = np.random.permutation(np.arange(len(agents)))

    # Let's do one step
    for agent_index1 in range(pop):
        agent_index2 = random_agents_order[agent_index1]
        # If we didn't find ourself do an exchange
        if agent_index1 != agent_index2:
            exchange =  exchange = random.random()

            agent_value1 = agents[agent_index1]
            agent_value2 = agents[agent_index2]

            # We have the same amount of stuff
            if agent_value1 == agent_value2:
                agent_value1 += agent_value2*exchange
                agent_value2 -= agent_value2*exchange

            # If agent1 has more than agent2, give a random amount of stuff to agent2 
            elif agent_value1 > agent_value2:
                agent_value2 += agent_value1*exchange
                agent_value1 -= agent_value1*exchange
            # Else agent2 has more than agent1, give a random amount of stuff to agent1
            else:
                agent_value1 += agent_value2*exchange
                agent_value2 -= agent_value2*exchange
            
            agents[agent_index1] = agent_value1
            agents[agent_index2] = agent_value2
    
    return agents
    
n, bins, patches = plt.hist(agents, 100)
plt.xlabel('Stuff')
plt.ylabel('Count')
plt.title('Initial Distribution')
plt.show()

iter_count = 1
iterations = [10, 100, 1000, 10000]
for each_iteration in iterations:
    while iter_count < each_iteration:
        agents = step(agents)
        iter_count += 1

    n, bins, patches = plt.hist(agents, 100)
    plt.xlabel('Stuff')
    plt.ylabel('Count')
    plt.title('After ' + str(iter_count) + ' iterations')
    plt.show()


```
